```python
Max_words = 100000  #these are the number of words in our vocab or the dictionary which are going to be tokenized i.e. mapped to numbers or integers.
```

```python
Vectorizer = TextVectorization(max_tokens =Max_words,
                               output_sequence_length = 1000,    #how long that comment's range go to 
                               output_mode = 'int')    #this specifies that each word is mapped to an integer
```

```python
vectorized_text = Vectorizer(X.values)    #its actually tokenizing the dataset
```

```python
#MCSHBAP - map , cache , shuffle ,batch ,prefetch 
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text , Y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)     #all four are data preprocessing steps. 
dataset = dataset.batch(12)
daatset = dataset.prefetch(8)     #helps bottlenecks
```

```python
 batch_X ,batch_y = dataset.as_numpy_iterator().next()
```

```python
int(len(dataset)*.7)
```

```python
train = dataset.take(int(len(dataset)*.7))         # .take method will basically grab or take that partition of data out and assign it to a variable .Here we have taken out 70 per. 6981 batches are being trained and convert it into an integer.
```

```python
validation_partition = dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2))    # we are going to skip 70 per of data and grab 20 per only.
```

```python
test_partition = dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*.1))    
```

```python
len(train)   # 6981 batches 
```

```python
train_generator = train.as_numpy_iterator()
train_generator.next()         #using this .next() , it is going to next batch .
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import LSTM , Dropout, Bidirectional , Dense , Embedding
```

```python
model = Sequential()      #this is going to instantiate the model. it is basically going to instantiate the sequential api.we can then add the number of layers to the model  to be abble to build it up.
model.add(Embedding(Max_words+1 , 32))     # creating the embedding layer.
```

```python
model.add(Bidirectional(LSTM( 32 , activation='tanh')))       #Bidirectional LSTM layer
model.add(Dense(128 , activation='relu' ))           
model.add(Dense(256 , activation='relu'))                 #Feature extractor or fully connected layers.
model.add(Dense(128 , activation='relu' ))
model.add(Dense(6 , activation = 'sigmoid'))            #Final layer.This final layer maps to the number of output that we have got inside of out neural network.
```

```python
model.compile(loss ='BinaryCrossentropy' , optimizer = 'Adam')
model.summary()
```

```python
history = model.fit( train , epochs= 6 ,validation_data = validation_partition)
```

```python
from tensorflow.keras.metrics import Precision ,Recall ,CategoricalAccuracy
```

```python
pre= Precision()         #here we have instantiated our precision metrics.
re = Recall()                          #similarly here also
acc = CategoricalAccuracy()            #here also
```

```python
for batch in test_partition.as_numpy_iterator():       #basically it loops through each and every single batch of our data pipeline 
     #unpacking the batch
     X_true , Y_true = batch
     
     #Make a prediction
     yhat = model.predict(X_true )
     
     #Flatten the predictions 
     Y_true = Y_true.flatten()
     yhat = yhat.flatten()
     
     pre.update_state(Y_true , yhat)
     re.update_state(Y_true , yhat)
     acc.update_state(Y_true , yhat)
    
```

```python
print(f'Precision : {pre.result().numpy()} , Recall:{re.result().numpy()} , Accuracy :{acc.result().numpy()}')
```

```python
def score_comment(comment):
    vectorized_comment = Vectorizer([comment])
    results = model.predict(vectorized_comment)
    
    text = ''
    for idx, col in enumerate(df.columns[2:]):
        text += '{}: {}\n'.format(col, results[0][idx]>0.5)
    
    return text
```

```python
interface = gr.Interface(fn=score_comment ,
                         inputs=gr.Textbox(lines=2, placeholder='Comment to score') , 
                        outputs='text')
```
